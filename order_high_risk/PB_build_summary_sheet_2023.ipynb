```python
# dispute --> Buyer report: Unauthorized --> Buyer report: Authorized --> Daily checking

daily_checking = summary_data[summary_data['check_reason'] == 'Daily Checking']
buyer_a = summary_data[summary_data['check_reason'] == 'Buyer report: Authorized']
buyer_u = summary_data[summary_data['check_reason'] == 'Buyer report: Unauthorized']
dispute_fraudulent = summary_data[summary_data['check_reason'] == 'Dispute fraudulent']
```

```python
buyer_u = buyer_u[~buyer_u['order_id'].isin(dispute_fraudulent['order_id'].tolist())]
df_1_2 = pd.concat([dispute_fraudulent, buyer_u])
```

```python
# lọc order buyer report: Authorized không nằm trong dispute & Buyer report: Unauthorized
buyer_a = buyer_a[~buyer_a['order_id'].isin(df_1_2['order_id'].tolist())]
df_1_3 = pd.concat([df_1_2, buyer_a])
```

```python
# lọc order Daily checking không nằm trong dispute & Buyer report: Unauthorized & buyer report: Authorized
daily_checking = daily_checking[~daily_checking['order_id'].isin(df_1_3['order_id'].tolist())]
```

```python
final_summary_data = pd.concat([df_1_3, daily_checking])
final_summary_data = final_summary_data.sort_values('create_at', ascending=False)
```

```python
final_summary_data = final_summary_data.sort_values(['create_at', 'final_action'], ascending=False).drop_duplicates('order_id', keep='first')
```

```python
final_summary_data.loc[-0.5] = summary_new_row
final_summary_data = final_summary_data.sort_index().reset_index(drop=True)
```

```python
final_summary_data.drop_duplicates(inplace=True)
final_summary_data.fillna('', inplace=True)
```
